# Plotting the 3-dB coupling length against the Coupler width

```python
# Directory containing the CSV files
data_dir = r"C:\Users\phill\OneDrive\Skole\Kandidat\1.Semester\Photonic Integrated Circuit Design\Projects\My Project\optical_couplers\simulation_data\MMI_COUPLER_data"

# Wavelengths to plot
wavelengths = [1500, 1525, 1550, 1575, 1600]

# Distinct line styles and markers for clarity
line_styles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1))]
markers = ['o', 's', 'D', '^', 'v']

plt.figure(figsize=(3.1, 2.2))

for i, wl in enumerate(wavelengths):
    filename = f"MMI_3dB_results_{wl}nm.csv"
    filepath = os.path.join(data_dir, filename)

    if not os.path.exists(filepath):
        print(f" File not found: {filename}")
        continue

    df = pd.read_csv(filepath)
    width_array = df["Width_m"].to_numpy()
    series_array = df["L3dB_from_n_m"].to_numpy()

    plt.plot(
        width_array,
        series_array,
        linestyle=line_styles[i % len(line_styles)],
        marker=markers[i % len(markers)],
        color='black',
        markersize=3,
        linewidth=0.9,
        label=fr"$\lambda = {wl}\,\mathrm{{nm}}$"
    )

plt.xlabel("Width [m]", fontsize=8)
plt.ylabel("3-dB Coupling Length [m]", fontsize=8)
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)
plt.grid(True, alpha=0.25, linewidth=0.4)
plt.legend(frameon=False, fontsize=7, handlelength=2)
plt.title("3-dB Coupling Length vs Width")
plt.tight_layout(pad=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(["science", "ieee"])

# ==============================
# Load Monte-Carlo data
# ==============================
DATA_PATH = r"C:\Users\phill\OneDrive\Skole\Kandidat\1.Semester\Photonic Integrated Circuit Design\Projects\My Project\optical_couplers\simulation_data\mmi_montecarlo_eta.npz"
data = np.load(DATA_PATH)

k   = data["eta"]
N_SAMPLES = len(k)

print("Loaded:", DATA_PATH)
print(f"Samples = {N_SAMPLES}")

# ==============================
# Histogram of k
# ==============================

mean_val = np.mean(k)
std_val  = np.std(k)

plt.figure(figsize=(6,4))
plt.hist(
    k,
    bins=40,
    edgecolor="grey",       # DEFINES BIN LINES
    linewidth=0.7,           # ensures visible separation
)

plt.xlabel(r"$k$ (cross-port power)")
plt.ylabel(r"Count")
plt.title(rf"Directional Coupler Monte Carlo ($N={N_SAMPLES}$)")

# Legend with mean + std
plt.legend(
    [rf"$\mu = {mean_val:.3f}$" + "\n" + rf"$\sigma = {std_val:.3f}$"],
    frameon=True,
)

plt.tight_layout()
plt.show()
```
